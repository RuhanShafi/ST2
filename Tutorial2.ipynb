```python
import random

def has_duplicate_birthdays(n):
    birthdays = []
    for _ in range(n):
        day = random.randint(1, 365)
        if day in birthdays:
            return True
        birthdays.append(day)
    return False

def birthday_experiment(trials=1000):
    for n in range(5, 101, 5):
        duplicate_count = 0
        for _ in range(trials):
            if has_duplicate_birthdays(n):
                duplicate_count += 1
        probability = duplicate_count / trials
        print(f"n = {n:3d}, Probability of shared birthday ≈ {probability:.3f}")

birthday_experiment()
```
